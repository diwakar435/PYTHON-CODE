# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1 = np.linspace(0,10,100)
fig = plt.figure()  # create a plot figure
plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--');
plt.show()
```

# Pyplot API

```python
plt.figure()  # create a plot figure

# create a first of two panels and set a current axis 
plt.subplot(2,1,1)  # (rows, columns, and panels)
plt.plot(x1,np.sin(x1))

# create a second of two panels and set a current axis
plt.subplot(2,1,2)
plt.plot(x1,np.cos(x1));
plt.show()
```

```python
# get current figure information
print(plt.gcf())
plt.show()
```

```python
# get current axis information
print(plt.gca())
plt.show()
```

# Visualization with pyplot

# plot() - Versatile command 

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

# State-machine interface

```python
x = np.linspace(0,2,100)

plt.plot(x, x, label = 'linear')
plt.plot(x, x**2, label = 'quadratic')
plt.plot(x, x**3, label = 'cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title('simple plot')

plt.legend()
plt.show()
```

# formatting the style of plot

```python
plt.plot([1,2,3,4],[1,4,9,16],'go')
plt.axis([0,6,0,20])
plt.show()
```

# Working with numpy arrays

```python
# evenly sampled time at 200ms intervals
t = np.arange(0.,5.,0.2)

# red dashes, blue squares and green triangles
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

# Object-oriented API

```python
# First create a grid of plots
# ax will be an array of two axes objects
fig,ax = plt.subplots(2)

# call a plot() method on the appropriate object 
ax[0].plot(x1,np.sin(x1),'b-')
ax[1].plot(x1,np.cos(x1),'b-');
plt.show()
```

# object and reference

```python
fig = plt.figure()

x2 = np.linspace(0,5,10)
y2 = x2**2

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x2,y2,'r')
axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title');
plt.show()
```

# Error bar chart

```python
x9 = np.arange(0,4,0.2)
y9 = np.exp(-x9)
e1 = 0.1 * np.abs(np.random.randn(len(y9)))

plt.errorbar(x9,y9,yerr = e1, fmt = '.-')
plt.show()
```

# Stacked bar chart

```python
A = [15.,30.,45.,22.]
B = [15.,25.,50.,20.]
z2 = range(4)
plt.bar(z2,A, color = 'b')
plt.bar(z2,B, color = 'r',bottom = A)
plt.show()
```

# Adding a legend

```python
x16 = np.arange(1,5)
fig,ax = plt.subplots()
ax.plot(x16, x16*1.5)
ax.plot(x16, x16*3.0)
ax.plot(x16, x16/3.0)
plt.legend(['fast','Slow','Medium'])
plt.show()
```

```python
x16 = np.arange(1,5)
fig,ax = plt.subplots()
ax.plot(x16, x16*1.5,label = 'fast')
ax.plot(x16, x16*3.0,label = 'Slow')
ax.plot(x16, x16/3.0,label = 'Medium')
plt.legend()
plt.show()
```
